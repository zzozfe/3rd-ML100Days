### 用 digits (dataset.load_digits()) 資料嘗試跑 t-SNE

```python
from sklearn import datasets

digits = datasets.load_digits()
X = digits.data
y = digits.target
```

```python
from sklearn import manifold
```

```python
n_samples, n_features = X.shape
n_neighbors = 30
tsne = manifold.TSNE(n_components=2, random_state=0, init='pca', learning_rate=100., early_exaggeration=12., perplexity=10)
```

```python
xx = tsne.fit_transform(X)
```

```python
import matplotlib.pyplot as plt
plt.scatter(xx[:,0], xx[:,1])
plt.show()
```
